```python
df1 = (
    {
        "first"  : ["Corey", "Jane","John"],
        "last" : ["Schafer", "Doe", "Doe"],
        'email': ["CoreySchafer@gmail.com", "JaneDoe@gmail.com", "JohnDoe@gmail.com"]
    }
)
df1 = pd.DataFrame(df1)
df1
```

```python
# Advanced Filter
# how to find the email whose last name is schafer or first name is John
filt = (df1['last'] == "Schafer") | (df1['first'] == 'John')
```

```python
df1.loc[filt, 'email']
```

```python
# Find the name whose last and first name is not John the oposite of first example
# for this we just use the the ~ negate
df1.loc[~filt, 'email']
# This is the name whose last name is not schafer or first name is John
```
